```python
def heuristic(node, goal):
    if node in Coordinates and goal in Coordinates:
        return great_circle(Coordinates[node], Coordinates[goal]).kilometers
    return float('inf')

heuristic("ICOR", "Central Hospital")
```

```python
import heapq

def a_star(graph, start, goal):
    distances = {node: float('inf') for node in graph}
    distances[start] = 0
    previous = {node: None for node in graph}  
    
    pq = [(0 + heuristic(start, goal), start)]
    
    while pq:
        current_f, current_node = heapq.heappop(pq)
        current_g = distances[current_node]
        
        if current_f > current_g + heuristic(current_node, goal):
            continue
        
        if current_node == goal:
            path = []
            while current_node:
                path.append(current_node)
                current_node = previous[current_node]
            path.reverse()
            return current_g, path
        
        for neighbor, weight in graph[current_node]:
            new_g = current_g + float(weight)
            if new_g < distances[neighbor]:
                distances[neighbor] = new_g
                previous[neighbor] = current_node
                new_f = new_g + heuristic(neighbor, goal)
                heapq.heappush(pq, (new_f, neighbor))
    
    return distances[goal], []
```

```python
dist, path = a_star(Map, "ICOR", "Central Hospital")
print(f"Distance: {dist:.2f} km")
print(f"Path: {' → '.join(path)}")
```

```python
def find_Hospital(graph, start):
    Hospitals = ["Central Hospital", "Military Hospital", "Private Hospital Lenmed", "Sommerschield Clinic", "ICOR"]
    closest_hospital = None
    for hospital in Hospitals:
        distance, path = a_star(graph, start, hospital)
        if closest_hospital is None or distance < closest_hospital[0]:
            closest_hospital = (distance, hospital, path)
    return closest_hospital
```
