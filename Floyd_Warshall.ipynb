```python
def floyd_warshall(Map):
      
        distances = {node: {neighbor: float('inf') for neighbor in Map} for node in Map}
        
        for node, neighbors in Map.items():
            distances[node][node] = 0
            for neighbor, weight in neighbors:
                distances[node][neighbor] = weight
        
       
        for k in Map:
            for i in Map:
                for j in Map:
                        if distances[i][j] > distances[i][k] + distances[k][j]:
                            distances[i][j] = distances[i][k] + distances[k][j]
        
        return distances
```

```python
import time

iterations = 10000
start_time = time.perf_counter()
for _ in range(iterations):
    floyd_warshall(Map)
end_time = time.perf_counter()

avg_time = (end_time - start_time) / iterations * 10000
print(f"Average Floyd-Warshall runtime: {avg_time:.4f} ms")
```
